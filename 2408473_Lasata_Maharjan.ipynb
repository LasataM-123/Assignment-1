**Task - 3 - Outlier Detection:**

**Tasks:**

1. Identify outlier countries in South Asia based on their Score and GDP per Capita.
2. Define outliers using the 1.5 Ã— IQR rule.
3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
in a different color.
4. Discuss the characteristics of these outliers and their potential impact on regional averages.

```python
#For score column,
#Finding quartile 1 (Q1)
Q1_score=SouthAsia['score'].quantile(0.25)

#Finding quartile 3 (Q3)
Q3_score=SouthAsia['score'].quantile(0.75)

#Finding interquartile range
IQR_score=Q3_score-Q1_score

#Finding the lower and upper outliers
outlier_score_lower=Q1_score-1.5*IQR_score
outlier_score_upper=Q3_score+1.5*IQR_score

#Finding the rows that meet the outliers' condition
score_outliers=SouthAsia[(SouthAsia['score']<outlier_score_lower) | (SouthAsia['score']>outlier_score_upper)]
print("Outliers in score column = \n")
score_outliers.head()
```

```python
#For Log GDP per capita column,
#Finding quartile 1 (Q1)
Q1_gdp=SouthAsia['Log GDP per capita'].quantile(0.25)

#Finding quartile 3 (Q3)
Q3_gdp=SouthAsia['Log GDP per capita'].quantile(0.75)

#Finding interquartile range
IQR_gdp=Q3_gdp-Q1_gdp

#Finding the lower and upper outliers
outlier_gdp_lower=Q1_gdp-1.5*IQR_gdp
outlier_gdp_upper=Q3_gdp+1.5*IQR_gdp

#Finding the rows that meet the outliers' condition
gdp_outliers=SouthAsia[(SouthAsia['Log GDP per capita']<outlier_gdp_lower) | (SouthAsia['Log GDP per capita']>outlier_gdp_upper)]
print("Outliers in Log GDP per capita column = \n")
gdp_outliers.head()
```

```python
#Storing the columns for reusability
gdp_column=SouthAsia['Log GDP per capita']
score_column=SouthAsia['score']
```

```python
plt.figure(figsize=(10,6))
#Scatter plot for normal points(non outliers)
non_outliers= (
    (gdp_column >= outlier_gdp_lower) & (gdp_column <= outlier_gdp_upper) &
    (score_column >= outlier_score_lower) & (score_column <= outlier_score_upper)
)
plt.scatter(
    gdp_column[non_outliers],
    score_column[non_outliers],
    color='black', label='Non-outliers'
)

#Scatter plot for Lower outliers
lower_outliers= (
    (gdp_column < outlier_gdp_lower) | (score_column < outlier_score_lower)
)
plt.scatter(
    gdp_column[lower_outliers],
    score_column[lower_outliers],
    color='red', label='Lower Outliers'
)

#Scatter plot for Upper outliers
upper_outliers = (
    (gdp_column > outlier_gdp_upper) | (score_column > outlier_score_upper)
)
plt.scatter(
    gdp_column[upper_outliers],
    score_column[upper_outliers],
    color='green', label='Upper Outliers'
)

# Adding legend, xlabel, ylabel and title
plt.legend()
plt.xlabel('Log GDP per capita')
plt.ylabel('Score')
plt.title('Scatter Plot of Non-Outliers and Outliers')
plt.show()
```

4 - > Observing all the calculations and the scatter plot, Afghanistan seems to have the lowest score and Log GDP per capita which represents economic underdevelopment compared to other 4 South Asian countries as well as it has low values in some related indicators such as social support, life expectancy, etc.

Due to the lower score and Log GDP per capita of Afghanistan, this may skew the regional averages downward which misinterprets the overall social progress and economic development of South Asia.

**Tasks:**

1. Choose two metrics (e.g., Freedom to Make Life Choices and Generosity) and calculate their correlation
{pearson correlation} with the Score for South Asian countries.
2. Create scatter plots with trendlines for these metrics against the Score.
3. Identify and discuss the strongest and weakest relationships between these metrics and the Score for
South Asian countries.

```python
#Q.no.1
#Choosing two metrics and score column
metric_1=SouthAsia['Freedom to make life choices']
metric_2=SouthAsia['Generosity']
score=SouthAsia['score']
```

```python
#Calculating the mean of score column
score_mean=score.mean()
```

```python
#Declaring a function for better code reusability
def correlation_coefficient(X,Y):
  '''
  This function calculates the pearson correlation coeeficient between two columns.
  Parameters:
  X: First column passed by the user
  Y: Second column passed by the user
  Returns:
  correlation_coefficient: Pearson correlation coefficient between X and Y
  '''

  #Calculating the mean of X and Y
  X_mean=X.mean()
  Y_mean=Y.mean()

  #Calculating the numerator
  numerator=sum((X-X_mean)*(Y-Y_mean))

  #Calculating the denominator
  denominator=(sum((X-X_mean)**2)*sum((Y-Y_mean)**2))**0.5

  #Calculating the correlation coefficient
  correlation_coefficient=numerator/denominator

  #returning the correlation coefficient
  return correlation_coefficient
```

```python
print("Pearson correlation coefficient of metric_1 with score = ",format(correlation_coefficient(metric_1,score),'.3f'))
print("Pearson correlation coefficient of metric_2 with score = ",format(correlation_coefficient(metric_2,score),'.3f'))
```

```python
#Q.no.2
#Scatter plot with trendline for metric_1
plt.figure(figsize=(8,6))
sns.regplot(x=metric_1, y=score, data=df, ci=None, scatter_kws={'color': 'blue','s': 50}, line_kws={'color': 'red','label':'trendline'})
plt.xlabel("Freedom to make life choices")
plt.ylabel("Score")
plt.title("Scatter plot of Freedom to make life choices with score")
plt.legend()
plt.show()
```
